```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
airline_data= pd.read_excel('EastWestAirlines (1).xlsx',sheet_name='data')
airline_data
```

```python
airline_data.rename(columns={'ID#':'ID','Award?':'Award'},inplace=True)
airline_data
```

```python
airline_data.describe()
```

```python
airline_data.info()
```

```python
from sklearn.preprocessing import normalize
```

```python
# Normalize heterogenous numerical data 
airline_norm=pd.DataFrame(normalize(airline_data),columns=airline_data.columns)
airline_norm
```

```python
airline_data.set_index('ID',inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
for feature in airline_data.columns:
    airline_data1=airline_data.copy()
    airline_data1[feature].hist(bins=25)
    plt.title('count')
    plt.title(feature)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(airline_data)
scaled_df[:5]
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
```

```python
model.fit(scaled_df)
```

```python
model.labels_
```

```python
airline_data['Labels']=model.labels_
```

```python
grp = airline_data.groupby('Labels')
```

```python
for x,y in grp:
    print(y)
    print('_'*40)
```

```python
grp.agg('mean')
```

```python
from scipy.cluster import hierarchy
import scipy.cluster.hierarchy as sch
```

```python
ln = hierarchy.linkage(scaled_df,method='complete')
```

```python
dn= hierarchy.dendrogram(ln)
plt.figure(figsize =(15,10))
```

```python
for methods in ['single','complete','average','weighted','centroid','median','ward']: 
    plt.figure(figsize =(14,6))    
    dict = {'fontsize':16,'fontweight' :14, 'color' : 'blue'}    
    plt.title('Visualising the data, Method- {}'.format(methods),fontdict = dict) 
    Dendrogram1 = sch.dendrogram(sch.linkage(scaled_df, method = methods,optimal_ordering=False))
```

```python
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
airline_data['clustersid']=hclusters.labels_
airline_data
```

```python
airline_data.groupby('clustersid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(airline_data['clustersid'],airline_data['Balance'], c=hclusters.labels_) 
```
